```python
import pandas as pd
from pandas import read_csv
import numpy as np
import matplotlib.pyplot as plt

#,           2017-2018 
dataset = pd.read_csv('input/data_2017-2018.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)

# 
atn = dataset.attendance
dri = dataset.drinks
des = dataset.desserts
firs = dataset.firstfood
sec = dataset.secondfood
sal = dataset.salads

#  2017-2018 
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(atn, label=r"attendance")
ax.set_xlabel(r'$Date$', fontsize=18)
ax.set_ylabel(r'$Attendance$', fontsize=18)
ax.set_title('Attendance 2017 - 2018')
ax.legend(loc=2);
fig.savefig('output/png/attendance_2017-2018.png')
fig.savefig('static/attendance_2017-2018.png')

#    2017-2018 
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(firs, label=r"firstfood")
ax.plot(sec, label=r"secondfood")
ax.plot(sal, label=r"salads")
ax.plot(dri, label=r"drinks")
ax.plot(des, label=r"desserts")
ax.set_xlabel(r'$Date$', fontsize=18)
ax.set_ylabel(r'$Sales$', fontsize=18)
ax.set_title('Statistics 2017 - 2018')
ax.legend(loc=2);
fig.savefig('output/png/sales_2017-2018.png')
fig.savefig('static/sales_2017-2018.png')

#   
k_dri = (dri)/atn
k_des = (des)/atn
k_firs = (firs)/atn
k_sec = (sec)/atn
k_sal = (sal)/atn

# 
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(k_firs, label=r"firstfood")
ax.plot(k_sec, label=r"secondfood")
ax.plot(k_sal, label=r"salads")
ax.plot(k_dri, label=r"drinks")
ax.plot(k_des, label=r"desserts")
ax.set_xlabel(r'$Date$', fontsize=18)
ax.set_ylabel(r'$Ratio$', fontsize=18)
ax.set_title('Ratio of sales 2017 -2018')
ax.legend(loc=2);
fig.savefig('output/png/ratio.png')
fig.savefig('static/ratio.png')

#,        
dataset_2007_2008 = read_csv('input/attendance_2007-2008.csv',',',index_col=False)
date = dataset_2007_2008.date

#        2007  2018 
dataset_2007_2008 = read_csv('input/attendance_2007-2008.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2007_2008 = np.array(dataset_2007_2008.attendance)
dataset_2008_2009 = read_csv('input/attendance_2008-2009.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2008_2009 = np.array(dataset_2008_2009.attendance)
dataset_2009_2010 = read_csv('input/attendance_2009-2010.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2009_2010 = np.array(dataset_2009_2010.attendance)
dataset_2010_2011 = read_csv('input/attendance_2010-2011.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2010_2011 = np.array(dataset_2010_2011.attendance)
dataset_2011_2012 = read_csv('input/attendance_2011-2012.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2011_2012 = np.array(dataset_2011_2012.attendance)
dataset_2012_2013 = read_csv('input/attendance_2012-2013.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2012_2013 = np.array(dataset_2012_2013.attendance)
dataset_2013_2014 = read_csv('input/attendance_2013-2014.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2013_2014 = np.array(dataset_2013_2014.attendance)
dataset_2014_2015 = read_csv('input/attendance_2014-2015.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2014_2015 = np.array(dataset_2014_2015.attendance)
dataset_2015_2016 = read_csv('input/attendance_2015-2016.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2015_2016 = np.array(dataset_2015_2016.attendance)
dataset_2016_2017 = read_csv('input/attendance_2016-2017.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2016_2017 = np.array(dataset_2016_2017.attendance)
dataset_2017_2018 = read_csv('input/data_2017-2018.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn_2017_2018 = np.array(dataset_2017_2018.attendance)

#   2018-2019 
prediction_2018_2019 = np.ceil((atn_2007_2008 + atn_2008_2009 + atn_2009_2010 
+ atn_2010_2011 + atn_2011_2012 + atn_2012_2013 + atn_2013_2014 + atn_2014_2015
+ atn_2015_2016 + atn_2016_2017 - atn_2007_2008 - atn_2007_2008 - atn_2007_2008 - atn_2007_2008 - atn_2007_2008
- atn_2007_2008 - atn_2007_2008 - atn_2007_2008 - atn_2007_2008 - atn_2007_2008 - atn_2007_2008)/11 + atn_2017_2018).astype(np.int64)
atn = pd.DataFrame(prediction_2018_2019)

#
Frame = [date, atn]
prediction1 = pd.concat(Frame, axis=1)
prediction1.columns = ['date', 'prediction_attendance']
prediction1.to_csv('output/prediction_attendance.csv')

# csv    
prediction_new = read_csv('output/prediction_attendance.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
atn2 = prediction_new.prediction_attendance

#   
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(atn2, label=r"attendance")
ax.set_xlabel(r'$Date$', fontsize=18)
ax.set_ylabel(r'$Prediction Attendance$', fontsize=18)
ax.set_title('Prediction Attendance')
ax.legend(loc=2);
fig.savefig('output/png/prediction_attendance.png')
fig.savefig('static/prediction_attendance.png')

#        ( - ),   
dataset_help = pd.read_csv('output/prediction_attendance.csv',',')
date = dataset_help.date

#          
#  DataFrame   2018-2019 ,  (np.ceil)      (astype(np.int64))
p_dri=pd.DataFrame(np.ceil(k_dri.values*atn2.values).astype(np.int64))
p_des=pd.DataFrame(np.ceil(k_des.values*atn2.values).astype(np.int64))
p_firs=pd.DataFrame(np.ceil(k_firs.values*atn2.values).astype(np.int64))
p_sec=pd.DataFrame(np.ceil(k_sec.values*atn2.values).astype(np.int64))
p_sal=pd.DataFrame(np.ceil(k_sal.values*atn2.values).astype(np.int64))

#    ,    csv
Frame = [date, p_dri, p_des, p_firs, p_sec, p_sal]
prediction = pd.concat(Frame, axis=1)
prediction.columns = ['date', 'drinks', 'desserts', 'first_food', 'second_food', 'salads']
prediction.to_csv('input/tmp/prediction_sales_temp.csv',index=False)

#   -   
dataset_2018_2019 = pd.read_csv('input/tmp/prediction_sales_temp.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
dataset_2018_2019.to_csv('output/prediction_sales.csv',index=False)

#     
p_dri = dataset_2018_2019.drinks
p_des = dataset_2018_2019.desserts
p_firs = dataset_2018_2019.first_food
p_sec = dataset_2018_2019.second_food
p_sal = dataset_2018_2019.salads

#     2018-2019 
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(p_firs, label=r"firstfood")
ax.plot(p_sec, label=r"secondfood")
ax.plot(p_sal, label=r"salads")
ax.plot(p_dri, label=r"drinks")
ax.plot(p_des, label=r"desserts")
ax.set_xlabel(r'$Date$', fontsize=18)
ax.set_ylabel(r'$Sales$', fontsize=18)
ax.set_title('Prediction Statistics 2018-2019')
ax.legend(loc=2);
fig.savefig('output/png/prediction_sales.png')
fig.savefig('static/prediction_sales.png')

#  txt      
itog = atn2.describe()
itog.to_csv('output/analyze.txt', sep='-')

#p - prediction ()
#r - revenue ()
#c - cost ()
prices = pd.read_csv('input/prices.csv',',')
c_dri = prices.drinks
c_des = prices.desserts
c_firs = prices.firstfood
c_sec = prices.secondfood
c_sal = prices.salads

#  
r_dri=pd.DataFrame(np.ceil(c_dri.values*p_dri.values).astype(np.int64))
r_des=pd.DataFrame(np.ceil(c_des.values*p_des.values).astype(np.int64))
r_firs=pd.DataFrame(np.ceil(c_firs.values*p_firs.values).astype(np.int64))
r_sec=pd.DataFrame(np.ceil(c_sec.values*p_sec.values).astype(np.int64))
r_sal=pd.DataFrame(np.ceil(c_sal.values*p_sal.values).astype(np.int64))
r_all=pd.DataFrame(r_dri+r_des+r_firs+r_sec+r_sal)

#   2017-2018 
r_dri_old=pd.DataFrame(np.ceil(c_dri.values*dri.values).astype(np.int64))
r_des_old=pd.DataFrame(np.ceil(c_des.values*des.values).astype(np.int64))
r_firs_old=pd.DataFrame(np.ceil(c_firs.values*firs.values).astype(np.int64))
r_sec_old=pd.DataFrame(np.ceil(c_sec.values*sec.values).astype(np.int64))
r_sal_old=pd.DataFrame(np.ceil(c_sal.values*sal.values).astype(np.int64))
r_all_old=pd.DataFrame(r_dri_old+r_des_old+r_firs_old+r_sec_old+r_sal_old)

#    ,    csv
Frame = [date, r_dri, r_des, r_firs, r_sec, r_sal, r_all]
prediction = pd.concat(Frame, axis=1)
prediction.columns = ['date', 'drinks', 'desserts', 'first_food', 'second_food', 'salads', 'all']
prediction.to_csv('input/tmp/prediction_temp2.csv',index=False)

#   -   
prediction_revenue = pd.read_csv('input/tmp/prediction_temp2.csv',',', index_col=['date'], parse_dates=['date'], dayfirst=True)
prediction_revenue.to_csv('output/prediction_revenue.csv',index=False)

#   
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(r_all, label=r"revenue")
ax.set_xlabel(r'$Date$', fontsize=18)
ax.set_ylabel(r'$Revenue$', fontsize=18)
ax.set_title('Prediction Revenue')
ax.legend(loc=2);
fig.savefig('output/png/prediction_revenue.png')
fig.savefig('static/prediction_revenue.png')

#     2017-2018  2018-2019 
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(r_all, label=r"2018-2019")
ax.plot(r_all_old, label=r"2017-2018")
ax.set_ylabel(r'$Revenue$', fontsize=18)
ax.set_title('Compare Revenue')
ax.legend(loc=2);
fig.savefig('output/png/compare_revenue.png')
fig.savefig('static/compare_revenue.png')

#    ()
dataset_2018_2019.head(10)
```
